```python
import numpy as np
import pandas as pd
import numpy.linalg as LA

SIZE = 25
```

```python
def norm(matr: list[list[float]]):
    """
    Реализация 6-й нормы
    """

    return 1 / SIZE * np.sqrt(sum(matr[i][j] ** 2 for j in range(SIZE) for i in range(SIZE)))


def cond(matr: list[list[float]]):
    """
    Вычисление числа обусловленности
    """

    return norm(matrix) * norm(LA.inv(matr))
```

```python
mass_matrix = [np.random.rand(SIZE, SIZE) for i in range(5)]
mdf = []
columns = ["Матрица", "Норма матрицы", "Число обусловленности"]
df = pd.DataFrame(columns=columns)
for matrix in mass_matrix:
    print(matrix, norm(matrix), cond(matrix))
    df.loc[len(df.index)] = [matrix, norm(matrix), cond(matrix)]
    mdf.append(pd.DataFrame(matrix))
df
```
